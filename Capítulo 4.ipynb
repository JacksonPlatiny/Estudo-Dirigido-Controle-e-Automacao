# Capítulo 4 - Características de sistema de controle com retroação:

O capítulo 4 inicia com a definição de um sistema de controle, relembrando que ele é uma interligação de componentes projetada para produzir uma saída desejada. Quando o comportamento desejado do sistema é conhecido, pode-se calcular o erro entre a saída real e a desejada, utilizando esse erro para gerar um sinal de controle. O autor introduz a ideia de sistemas de malha fechada, também chamados de sistemas com realimentação, que utilizam essa estratégia para ajustar continuamente a saída e minimizar desvios. O foco do capítulo é demonstrar como a realimentação influencia as características de desempenho de um sistema de controle, melhorando sua estabilidade, reduzindo erros e tornando-o mais robusto a perturbações e variações dos parâmetros.

A realimentação é uma técnica fundamental no controle de sistemas dinâmicos, permitindo que eles reajam de maneira mais eficiente a mudanças no ambiente. O autor apresenta um exemplo biológico para ilustrar a importância da realimentação: o sistema de controle do ritmo cardíaco humano. Esse sistema ajusta continuamente os batimentos cardíacos de acordo com a necessidade de oxigenação do corpo, de forma análoga ao que ocorre em um sistema de controle automatizado. Em seguida, o autor discute a diferença entre sistemas de malha aberta e malha fechada. Os sistemas de malha aberta operam sem realimentação, produzindo uma saída em resposta direta a um sinal de entrada, sem considerar as variações do ambiente. Já os sistemas de malha fechada utilizam sensores para medir a saída e compará-la com a referência desejada, gerando um sinal de erro que corrige o comportamento do sistema.

O capítulo aborda a relação matemática entre a função de transferência do sistema e o erro resultante. O autor demonstra que, para reduzir o erro, a magnitude da expressão 1 + GH(s) deve ser maior que 1 na faixa de valores de s  analisada. Isso significa que a presença da função de realimentação H(s) pode reduzir significativamente o erro de rastreamento, tornando o sistema mais preciso. Quando H(s) se aproxima de 1, a estimativa do erro se torna mais precisa, permitindo um controle mais efetivo.

Outro ponto fundamental abordado é a sensibilidade do sistema às variações nos parâmetros. Em sistemas reais, fatores como mudanças ambientais, envelhecimento dos componentes e imprecisões nos parâmetros podem afetar significativamente a resposta do sistema. Em um sistema de malha aberta, qualquer variação nos parâmetros pode resultar em grandes desvios na saída, pois não há correção automática. No entanto, em sistemas de malha fechada, a realimentação atua para reduzir essa sensibilidade, ajustando continuamente a entrada de forma a compensar as variações nos parâmetros. O autor define a sensibilidade do sistema como a relação percentual entre a variação da função de transferência e a variação dos parâmetros do processo. O capítulo destaca que essa propriedade é especialmente útil em aplicações como amplificadores eletrônicos na indústria de telecomunicações, onde variações nos componentes podem comprometer a qualidade do sinal.

A resposta transitória do sistema também recebe grande atenção. O autor explica que a resposta transitória corresponde ao comportamento do sistema antes de atingir o regime permanente. A realimentação pode ser utilizada para controlar características essenciais da resposta transitória, como o tempo de subida (tempo necessário para atingir o valor desejado pela primeira vez), o overshoot (quanto a resposta ultrapassa o valor desejado antes de estabilizar) e o tempo de acomodação (tempo necessário para a resposta se manter dentro de uma faixa especificada em torno do valor final). Por meio da realimentação, os polos do sistema podem ser reposicionados no plano complexo, alterando diretamente a velocidade e o amortecimento da resposta.

Outro aspecto crítico analisado no capítulo é o controle de perturbações. O autor define uma perturbação como qualquer entrada indesejada que afeta o sinal de saída do sistema. Nos sistemas de malha aberta, perturbações externas podem causar variações na saída sem que haja um mecanismo para corrigi-las. Já nos sistemas de malha fechada, a realimentação permite detectar e compensar essas perturbações automaticamente, tornando o sistema mais resistente a fatores externos, como oscilações de carga, mudanças de temperatura ou ruídos em sensores. O autor destaca que um dos principais motivos para introduzir realimentação nos sistemas de controle é justamente a capacidade de rejeição de perturbações, um fator crítico em aplicações industriais e automotivas.

Na sequência, o capítulo trata do erro de estado estacionário, que corresponde ao erro residual após a fase transitória. O autor explica que esse erro pode ser calculado utilizando o teorema do valor final e que, nos sistemas de malha aberta, o erro de estado estacionário tende a ser maior, pois não há um mecanismo de correção contínua. Em sistemas de malha fechada, o erro pode ser reduzido significativamente, especialmente quando se utilizam controladores com ação integral, como os controladores PID. A presença do termo integral no controlador garante que a saída seja continuamente ajustada para alcançar a referência desejada, eliminando o erro residual.

Porém, o capítulo também alerta para os custos e desafios da realimentação. Embora ela traga benefícios importantes, como maior precisão e rejeição de perturbações, sua implementação tem implicações práticas que devem ser consideradas. O autor destaca três desafios principais: (1) aumento da complexidade do sistema, já que projetar e implementar um sistema de malha fechada exige mais sensores, atuadores e processamento; (2) perda de ganho, pois a realimentação reduz o ganho efetivo do sistema em comparação a um sistema de malha aberta; e (3) risco de instabilidade, uma vez que um ganho muito elevado pode fazer com que o sistema oscile excessivamente ou até mesmo se torne instável.

Por fim, o capítulo apresenta exemplos práticos de sistemas que utilizam realimentação para melhorar seu desempenho. Um dos exemplos discutidos é o das máquinas perfuratrizes do Canal da Mancha, que enfrentavam desafios significativos devido à variação das condições do solo. Para garantir precisão na escavação, essas máquinas utilizavam um sistema de controle realimentado que ajustava continuamente a força de perfuração de acordo com as mudanças na resistência do solo. Outro exemplo importante abordado no capítulo é o controle dos rovers exploradores de Marte, que necessitam operar de forma autônoma em um ambiente incerto e hostil. Os sistemas de realimentação são essenciais para ajustar a navegação, controlar a velocidade e manipular os braços robóticos, permitindo que os rovers realizem suas missões com sucesso, mesmo diante de terrenos irregulares e variações de temperatura.

## Exemplo 4.1: Feedback amplifier

```python
import numpy as np
import matplotlib.pyplot as plt
import control as ctl

# Exemplo 4.1 - Amplificador com realimentação
Ka = 10000  # Ganho do amplificador
beta = 0.1   # Fator de realimentação

# Função de transferência do sistema sem realimentação
G_open = ctl.tf([-Ka], [1])

# Função de transferência do sistema com realimentação
G_closed = ctl.feedback(G_open, beta)

print("Função de transferência do sistema com realimentação:")
print(G_open)

print("Função de transferência do sistema com realimentação:")
print(G_closed)
```

Este exemplo ilustra como a realimentação pode reduzir a sensibilidade a variações nos parâmetros de um sistema.


*   Sistema sem realimentação: Um amplificador tem um ganho Ka, mas qualquer variação nesse valor pode afetar significativamente a saída.
*   Sistema com realimentação: A introdução de um fator de realimentação β reduz a sensibilidade do sistema às variações do ganho. A nova função de transferência do sistema fechado é: Gfechado(s)= −Ka / (1 + βKa)

Quanto maior for Ka, mais o ganho do sistema realimentado se torna independente de variações no amplificador, melhorando a robustez. Em outras palavras, a realimentação melhora a estabilidade e reduz a dependência dos componentes do sistema, tornando-o mais confiável.


## Exemplo 4.2: English Channel boring machines

```python
# Exemplo 4.2 - Controle da Máquina de Perfuração do Canal da Mancha
K = 100  # Ganho do controlador
num = [K + 1, 1]  # Numerador da FT

# Denominador do sistema
s = ctl.tf([1, 0], 1)
G_machine = ctl.tf(num, [1, 12, K])

time = np.linspace(0, 2, 1000)
td_response = ctl.step_response(G_machine, time)

plt.figure()
plt.plot(td_response.time, td_response.outputs, label="Resposta ao Degrau")
plt.xlabel("Tempo (s)")
plt.ylabel("Ângulo da Máquina")
plt.title("Resposta ao Degrau - Máquina de Perfuração")
plt.legend()
plt.grid()
plt.show()
```

Este exemplo analisa o efeito da realimentação em um sistema de controle mecânico utilizado para escavar o túnel sob o Canal da Mancha. Podemos destacar que:
*   O sistema de controle precisa garantir que a perfuração siga um ângulo desejado.
*   A modelagem resulta em uma função de transferência da forma: G(s) = (K + 1) / (s² + 12s + K).
*   A resposta ao degrau simula como a máquina reage a uma entrada de controle.
*   Com ganhos maiores, a resposta fica mais rápida e o sistema rejeita melhor perturbações.

Tendo em vista tudo isso, podemos afirmar que ajustar o ganho K afeta diretamente a precisão e estabilidade da perfuração. O uso de realimentação melhora o desempenho, reduzindo os desvios causados por distúrbios externos.




## Exemplo 4.3: Blood pressure control during anesthesia

```python
# Exemplo 4.3 - Controle da Pressão Arterial com PID
p = 2  # Parâmetro do paciente
Kp, Ki, Kd = 6, 4, 4  # Parâmetros do controlador PID

# Função de transferência do paciente
G_patient = ctl.tf([1], [1, 2*p, p**2])

# Controlador PID
G_pid = ctl.tf([Kd, Kp, Ki], [1, 0])

# Malha fechada
G_closed_pid = ctl.feedback(G_pid * G_patient)

# Simulação do degrau
t = np.linspace(0, 20, 1000)
response = ctl.step_response(G_closed_pid, t)

plt.figure()
plt.plot(response.time, response.outputs, label="Resposta ao Degrau")
plt.xlabel("Tempo (min)")
plt.ylabel("Pressão Arterial")
plt.title("Resposta ao Degrau - Controle de Pressão Arterial")
plt.legend()
plt.grid()
plt.show()
```

Já este exemplo explora como um controlador PID pode ser usado para regular a pressão arterial de um paciente sob anestesia. Podemos aferir que:

*   A pressão arterial média (MAP) é controlada ajustando a quantidade de anestesia administrada.
*   O modelo do paciente é representado por: G(s) = 1 / (s + 𝑝)², onde
𝑝 é um parâmetro fisiológico.
*   Um controlador PID com ganhos 𝐾𝑝, 𝐾𝑖 e 𝐾𝑑
  ajusta a resposta do sistema.
*   A resposta ao degrau permite verificar a estabilidade e precisão do controle.

Ou seja, podemos perceber que o PID reduz o erro em regime permanente e melhora a resposta transitória, porém, ajustes inadequados podem levar a oscilações excessivas ou resposta lenta.

# Recapitulação geral e pontos principais do capítulo:

## 4.1 Introdução

*   Um sistema de controle é um conjunto interconectado de componentes projetados para gerar uma resposta desejada.
*   A realimentação é usada para comparar a saída do sistema com a entrada desejada, gerando um sinal de erro que guia o controle.
*   Os sistemas de malha aberta não possuem realimentação e são sensíveis a distúrbios e variações dos parâmetros do processo.
*   Os sistemas de malha fechada (com realimentação) reduzem a sensibilidade a essas variações, melhoram a rejeição de distúrbios e reduzem o erro em regime permanente.

## 4.2 Análise do Sinal de Erro

*   Em um sistema de controle realimentado, o erro de rastreamento é definido como: E(s) = R(s) − Y(s), onde R(s) é a entrada desejada e Y(s) é a saída real.
*   A equação de erro pode ser reescrita usando a função de ganho de malha: L(s) = Gc(s) * G(s)
*   O erro de rastreamento é reduzido quando o ganho de malha L(s) é alto.
*   Existe um compromisso entre minimizar erro de rastreamento e atenuar ruído de medição.

## 4.3 Sensibilidade a Variações de Parâmetros


*   Em sistemas reais, os parâmetros do processo variam devido a mudanças ambientais e envelhecimento dos componentes.
*   Em malha aberta, essas variações impactam diretamente a saída.
*   Em malha fechada, a realimentação reduz essa sensibilidade, pois a resposta do sistema se torna menos dependente de G(s).
*   A função de sensibilidade é definida como: S(s)= 1 / (1+L(s)).
*   Quanto maior o ganho de malha L(s), menor a sensibilidade do sistema às variações de G(s).

## 4.4 Rejeição de Distúrbios

*   Muitos sistemas estão sujeitos a sinais de distúrbio externos, como ruído em sensores ou variações de carga.
*   A equação do erro pode ser escrita para incluir o distúrbio Td(s): E(s)= [1 / (1+L(s))] * R(s) − [G(s) / (1+L(s))] * Td(s).
*   Para minimizar os efeitos do distúrbio Td(s), deseja-se um L(s) alto em baixas frequências.
*   Para minimizar o impacto do ruído de medição N(s), deseja-se um L(s) baixo em altas frequências.
*   Apresentação do exemplo prático do controle de velocidade em um laminador de aço, onde um sistema realimentado reduz o impacto de variações súbitas de carga.

## 4.5 Controle da Resposta Transitória

*   A resposta transitória de um sistema afeta seu desempenho.
*   Um sistema de malha aberta pode ter tempos de resposta lentos ou comportamento oscilatório indesejado.
*   A realimentação pode ser usada para alterar a função de transferência e melhorar a resposta transitória.
*   Apresentação do exemplo prático de um sistema de controle de velocidade em motores pode ter tempo de resposta reduzido com realimentação e ajuste de ganho.

## 4.6 Erro em Regime Permanente


*   O erro de regime permanente é definido como o erro após a fase transitória.
*   Para um sistema realimentado com ganho DC alto, o erro em regime permanente pode ser reduzido significativamente.
*   O erro para entradas padrão (degrau, rampa e parábola) depende do tipo do sistema (número de integradores na malha).

## 4.7 O Custo da Realimentação


*   Embora a realimentação ofereça vantagens, também há custos associados:
  1.   Aumento da complexidade: mais sensores e atuadores necessários.
  2.   Perda de ganho: a realimentação reduz o ganho total do sistema.
  3.   Possibilidade de instabilidade: um sistema fechado pode se tornar instável dependendo do projeto do controlador.
